```python
from imutils import *
```

```python
image = imread('test.jpg')
image.shape
```

```python
(R,G,B) = cv2.split(image)
print(R.shape)
print(G.shape)
print(B.shape)
```

```python
merged = cv2.merge([R,G,B])
show(merged)
```

```python
cv2.imshow('R',R)
cv2.imshow('G',G)
cv2.imshow('B',B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
